```python
class With_Slots:
    __slots__ = ("name" , "age")

    def __init__(self, name, age):
        self.name = name
        self.age = age
```

```python
class Without_Slots:
    def __init__(self, name, surname):
        self.name = name
        self.surname = surname
```

```python
with_sl = With_Slots("Varazdat", 56)
```

```python
without_sl = Without_Slots("Narek", "Karapetyan")
```

```python
import time

start = time.perf_counter()
with_sl = With_Slots("Varazdat", 56)
end = time.perf_counter()

print(f"total time {(end - start):.5f}")
```

```python
start = time.perf_counter()
with_sl = Without_Slots("Varazdat", "Chibuxchyan")
end = time.perf_counter()

print(f"total time {(end - start):.5f}")
```

```python
class Student:
    s_id = 0  #class attribute
    def __init__(self, name, age):
        self.name = name  #instance attribute
        self.age = age
        self.id = Student.s_id 
        Student.s_id += 1
```

```python
s1 = Student("Alex", 15)
s2 = Student("Gabriel", 17)
s3 = Student("Karapet", 15)
```

```python
print(s1.id, s2.id, s3.id)
```

Abstract and Interface Classes
Abstract
1. Ունի գեթ 1 աբստրակտ մեթոդ
2․ Կարող է ունենալ ոչ աբստրակտ մեթոդներ և հատկություններ
3․ Պետք է ժառանգ լինին abc.ABC կլասից
4․ Պետք է կախված լինի abstractmethod դեկորատորը

Interface
1. Բոլոր մեթոդները աբստրակտ են
2․ Չի կարող ունենալ հատկություններ և ոչ աբստրակտ մեթոդներ
3․ Պետք է ժառանգ լինին abc.ABC կլասից
4․ Պետք է կախված լինի abstractmethod դեկորատորը


```python
from abc import ABC, abstractmethod

class Car(ABC):
    def __init__(self, make, model):
        self.make = make
        self.model = model

    @abstractmethod
    def drive(self):
        ...

    def description(self):
        print(self.make, self.model)
```

```python
class SuperCar(Car):
    def __init__(self, make, model):
        super().__init__(make, model)

    def drive(self):
        print("SuperCar driving...")

c1 = SuperCar("Porsche", "911")
```

```python
class Flyable(ABC):
    @abstractmethod
    def fly(self):
        ...

class Bird(Flyable):
    def fly(self):
        print("Flying ...")

b1 = Bird()
b1.fly()
```
