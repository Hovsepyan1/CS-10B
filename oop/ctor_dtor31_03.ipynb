```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    @property
    def name(self):  #getter
        return self.__name

    @name.setter
    def name(self, value):  #setter
        if not isinstance(value, str):
            raise TypeError("Invalid Type")
        if not value:
            raise ValueError("Invalid Value")
        self.__name = value

    @property
    def age(self): #getter
        return self.__age

    @age.setter
    def age(self, value): #setter
        if not isinstance(value, int):
            raise TypeError("Invalid Type")
        if value < 0:
            raise ValueError("Invalid Value")
        self.__age = value

    # def __del__(self): #destructor
    #     print("Deleting object")
    #     del self

    def __iter__(self):
        yield self.name
        yield self.age


        
```

```python
class List(list):
    def __iter__(self):
        for i in range(len(self) - 1, -1, -1):
            yield self[i]
```

```python
a = List([1,2,3])
for i in a:
    print(i, end = " ")
```

```python
class Person:
    def __new__(cls, *args, **kwargs):
        print("__new__ called")
        instance = object.__new__(cls)
        return instance

    def __init__(self, name, age):
        print("__init__ called")
        self.name = name
        self.age = age
```

```python
class Person:
    def __new__(cls, name, age):
        instance = object.__new__(cls)
        instance.get_age = lambda: age
        instance.get_name = lambda: name
        instance.name = name
        instance.age = age
        return instance
```

```python
p = Person("Eduard", 15)
```

```python
class Decorator:
    def __init__(self, func):
        self.func = func

    def __call__(self, *args, **kwargs):
        print("Before")
        x = self.func(*args, **kwargs)
        print("After")
        return x
@Decorator  #foo = Decorator(foo)
def foo():
    print("Foo")
    return 5

print(foo())
```

```python
class Person:
    __slots__ = ["age", "name"]
    def __init__(self, name, age):
        self.name = name
        self.age = age
```

```python
p = Person("Edik", 16, "Andreyan")
# p.__dict__
```

Փորձարկել կլասների ստեղծումը `__slots__` -ով և առանց, և՛ ժամանակային տեսանկյանունից և՛ հիշողության։
Փորձել կոտրել `__slots__`-ը, այսինքն `__slots__` գրելուց հետո կարողանանք ատրիբուտ ավելացնել

```python
class Student(Person):
    __slots__ = Person.__slots__ + ["gpa"]
    def __init__(self, name, age, gpa):
        super().__init__(name, age)
        self.gpa = gpa


s = Student("Gor", 15, 1.0)
# s.surname = "Danielyan"
```
