```python
from abc import ABC, abstractmethod
class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        print("Generic sound")

class Dog(Animal):
    def make_sound(self):
        print("Woof")

class Cat(Animal):
    def make_sound(self):
        print("Myau")

class Cow(Animal):
    def make_sound(self):
        print("Muu")

d = Dog()
c = Cat()
c1 = Cow()

d.make_sound()
c.make_sound()
c1.make_sound()
```

```python
l = [d, c, c1]
def make_sound(ls):
    for i in ls:
        i.make_sound()

make_sound(l)
```

```python
from functools import singledispatch

@singledispatch
def power(a):
    print(f"Processing value {a} with type {type(a)}")

@power.register
def _(a :int):
    print("Call int")
    return a ** 2

@power.register
def _(a :float):
    print("Call float")
    return a ** 2

print(power(4))
print(power(3.5))
```

```python
@singledispatch
def add(a, b):
    print(f"Processing values {a,b} with types {type(a), type(b)}")

@add.register
def _(a:int, b:int):
    print("Call int")
    return a + b

@add.register
def _(a:float, b:float):
    print("Call float")
    return a + b

@add.register
def _(a: int, b: float):
    print("Call int, float")
    return a + b

print(add(3, 4))
print(add(5.2, 2.6))
print(add(4, 5.3))
```

```python
from abc import ABC, abstractmethod
class Person(ABC):
    @abstractmethod
    def walk(self):
        print("Person walking...")

class Student:
    def walk(self):
        print("Student walking")

Person.register(Student)
s = Student()
print(isinstance(s, Person))
print(issubclass(Student, Person))
print(Student.mro())
```

```python
from typing import Protocol, runtime_checkable

@runtime_checkable
class Person(Protocol):
    def walk(self):
        ...

    def bar(self):
        ...

class Student:
    def walk(self):
       ...

    def foo(self):
        ...

s = Student()
print(isinstance(s, Person))
print(issubclass(Student, Person))
```

```python
class Student:
    school = "ACT"
    def __init__(self, name):
        self.name = name

    @classmethod
    def change_school(cls, new_school):
        print(cls)
        cls.school = new_school

s = Student("Taron")
print(s.school)
print(Student.school)
```
