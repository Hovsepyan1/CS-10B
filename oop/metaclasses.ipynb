```python
from datetime import datetime
class CustomType(type):
    def __new__(mcls, name, bases, cls_dict):
        print("Creating class")
        print(f"mcls = {mcls}")
        print(f"name = {name}")
        print(f"bases = {bases}")
        cls = super().__new__(mcls, name, bases, cls_dict)
        cls.age = lambda year: datetime.now().year - year
        return cls
```

```python
class Person(metaclass=CustomType):
    def __new__(cls, *args, **kwargs):
        ...
```

```python
print(Person.__dict__)
```

```python
Person.age(2000)
```

```python
class Singletone(type):
    def __new__(mcls, name, bases, cls_dict):
        ...

    def __call__(cls, *args, **kwargs):
        print(cls)
        print(args)
        print(kwargs)
        x = super().__call__(*args, **kwargs)
        return x
```

```python
class Person(metaclass=Singletone):
    def __init__(self, name, age):
        self.name = name
        self.age = age
```

```python
Person("Vardan", 15)
```

```python
class MSingletone(type):
    instances = {}
    def __call__(cls, *args, **kwargs):
        print("I HAVE BEEN CALLED (__call__)")
        print(f"i was called by {cls} and got args {args} qand kwargs {kwargs}")
        if cls in MSingletone.instances:
            return MSingletone.instances[cls]
        x = super().__call__(*args, **kwargs)
        MSingletone.instances[cls] = x
        return x



class test1(metaclass = MSingletone):
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"test({self.name}, {self.age})"
```

```python
a = test1("aaa", 15)
b = test1("bbb", 15)
```
