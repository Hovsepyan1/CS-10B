Ստեղծել հետևյալ կլասենրը, որոնք ունեն հետևյալ ատրիբուտները և մեթոդները՝
Student
name, grade
@classmethod from_string(cls, data) — ստանում է "Name-Grade", վերադարձնում Student օբյեկտ

```python
class Student:
    def __init__(self,name,grade):
        self.name=name
        self.grade=grade

    @classmethod
    def from_string(cls,data):
        name,grade=data.split("-")
        print(name,grade)
        return cls(name,int(grade))
    
    def __str__(self):
        return f"Student - {self.name},grade - {self.grade}"
    
    #def __repr__()
```

```python
s1=Student("Narek",16)
print(s1.from_string("Misak-16"))
```

Book
title, pages
@classmethod որը հետևում է քանի գիրք է ստեղծվել ընդհանուր (class-level counter)


```python
class Book:
    counter = 0
    def __init__(self, title , pages):
        self.title = title
        self.pages = pages
        Book.counter += 1


    @classmethod 
    def created_books_count(cls):
        print(cls.counter)
```

```python
b1 = Book("JS", 500)
b2 = Book("Python", 400)
b3 = Book("Java basics", 900)
b4 = Book("Cs basics", 1000)

Book.created_books_count()
```

MathUtils
Միայն @staticmethod-ներ, երեք մաթեմատիկական ստուգում՝
is_power_of_two(n) — արդյո՞ք 2-ի աստիճան է
is_perfect(n) — արդյո՞ք կատարյալ թիվ է
is_armstrong_number(n) — արդյո՞ք Արմսթրոնգի թիվ է

```python
class MathUtils:
    def __init__(self):
        pass

    @staticmethod
    def is_power_of_two(n):
        return n & (n-1) == 0
```

```python
print(MathUtils.is_power_of_two(7))
```
